```python
import simple_colors as color
```

```python
tasks = []

def add_task(task, due_date=None, priority=1):
    tasks.append({"task": task, "due_date": due_date, "priority": priority, "completed": False})
    print(color.green(f"\nTask '{task}' added to the to-do list."))

def remove_task(task_index):
    if 1 <= task_index <= len(tasks):
        removed_task = tasks.pop(task_index - 1)
        print(color.green(f"\nTask '{removed_task['task']}' removed from the to-do list."))
    else:
        print(color.red("\nInvalid task index."))

def mark_task_completed(task_index):
    if 1 <= task_index <= len(tasks):
        tasks[task_index - 1]["completed"] = True
        print(color.green("\nTask marked as completed."))
    else:
        print(color.red("\nInvalid task index."))

def show_tasks():
    if not tasks:
        print(color.yellow("\nNo tasks in the to-do list."))
    else:
        sorted_tasks = sorted(tasks, key=lambda x: x['priority'], reverse=True)
        print(color.green("\nSorted To-Do List (by Priority):"))
        for idx, task_info in enumerate(sorted_tasks, start=1):
            task = task_info['task']
            due_date = task_info['due_date']
            priority = task_info['priority']
            completed = task_info['completed']
            status = "Completed" if completed else "Pending"
            if due_date:
                print(color.blue(f"\n{idx}. {task} - Due: {due_date} \n Priority: {priority}  \n Status: {status}"))
            else:
                print(color.cyan(f"\n{idx}. {task} - Priority: {priority} - \n Status: {status}"))

def main():
    while True:
        print(color.green("\nSelect an option:",'bold'))
        print(color.blue("1. Add task"))
        print(color.blue("2. Remove task"))
        print(color.blue("3. Mark task as completed"))
        print(color.blue("4. Show tasks"))
        print(color.red("5. Exit"))

        choice = input("Enter your choice: ")

        if choice == '1':
              task = input("Enter the task: ")
              due_date = input("Enter due date (optional): ")
              priority = int(input("Enter priority (1 to 5, where 5 is highest): "))
              add_task(task, due_date, priority)
        elif choice == '2':
              show_tasks()
              task_index = int(input("Enter the task index to remove(from 0): "))
              remove_task(task_index)
        elif choice == '3':
              show_tasks()
              task_index = int(input("Enter the task index to mark as completed: "))
              mark_task_completed(task_index)
        elif choice == '4':
              show_tasks()
        elif choice == '5':
              print(color.green("Exiting the to-do list app.",'bold'))
              break
        else:
              print(color.red("Invalid choice. Please select a valid option."))

if __name__ == "__main__":
              print(color.green("\t\t\t\tT-D-L",['bold','underlined']))
              print(color.blue("\t\t\tSet your Priorities"))
              print(color.red("============================================================",'bold'))
              main()
```
